```python
X = df2.drop(columns=['mod_date','pub_date','cvss'])
y = df2.cvss
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# putting models in a dictionary
models = {"XGBoost Regressor": XGBRegressor(),
          "Random Forest": RandomForestRegressor(),
          "Catboost": CatBoostRegressor()         }

# setting up a function to fit and score model
def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
    fits and evaluates given machine learning models
    models: a dictionary of different scikit learn machine learning models
    X_train: training data (no labels)
    X_test: testing data(no labels)
    y_train: training labels
    y_test: testing labels
    """
    #setting up a random seed
    np.random.seed(42)
    
    # making a dictionary to keep model scores
    model_scores = {}
    
    #looping through models
    for name, model in models.items():
        #fitting the model to the data
        model.fit(X_train, y_train)
        # evaluating the model and appending its score to model_scores
        model_scores[name] = model.score(X_test, y_test)
    return model_scores
```

```python
fit_and_score(models,X_train,X_test,y_train,y_test)
```

```python
#building a custom function for fitting and predicting cvss
def fit_and_predict(data):
    import numpy as np
    import pandas as pd
    from catboost import CatBoostRegressor
    from sklearn.preprocessing import LabelEncoder
    from sklearn.model_selection import train_test_split
    from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
    import joblib
    data.pub_date = pd.to_datetime(data.pub_date)
    data.mod_date = pd.to_datetime(data.mod_date)
    X = data.drop(columns=['mod_date','pub_date','cvss'])
    y = data.cvss
    categ_col = data.select_dtypes(include=['object']).columns
    le = LabelEncoder()
    for i in categ_col:
        X[i] = le.fit_transform(X[i])
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
    cat = CatBoostRegressor()
    cat.fit(X_train,y_train)
    filename = 'finalized_model.sav'
    joblib.dump(cat,filename)
    y_pred = cat.predict(X_test)
    print('                              ')
    print('=============================')
    print(f'rmse==>{np.sqrt(mean_squared_error(y_test,y_pred))}')
    print(f'mae==>{mean_absolute_error(y_test,y_pred)}')
    print('=============================')

    cvss_pred = pd.DataFrame({'cvss_pred':y_pred})
    return cvss_pred

#building a custom function for a single data point
def single_data_point_pred(data):
    import numpy as np
    import pandas as pd
    from catboost import CatBoostRegressor
    from sklearn.preprocessing import LabelEncoder
    from sklearn.model_selection import train_test_split
    from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
    import joblib

    X = data.drop(['mod_date','pub_date'])
    loaded_model = joblib.load('finalized_model.sav')
    y_pred = loaded_model.predict(X)
    return y_pred
    
    

    

    
    
    

    

    
```
